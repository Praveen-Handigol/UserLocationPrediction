```python
def max_words(tweets):
  maxlen = 0
  for sentence in tweets:
    sentence = str(sentence)
    words = sentence.split()
    if len(words)>maxlen:
      maxlen = len(words)

  return maxlen
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_X_np, y_np, test_size=0.2, random_state=42)
```

```python
scale_columns = X_train[:,:49]
print(scale_columns[0])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scale_columns_after = scaler.fit_transform(scale_columns)
scale_columns_after[0]
```

```python
X_train[:, :49] = scale_columns_after
```

```python
scale_columns_test = X_test[:,:49]
scale_columns_after_test = scaler.transform(scale_columns_test)
scale_columns_after_test[0]
```

```python
X_test[:, :49] = scale_columns_after_test
```

```python
X_train.shape[1]
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define your ANN model
model = Sequential([
    Dense(649, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(2)  # Output layer with 2 neurons for latitude and longitude prediction
])

# Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Print the model summary
model.summary()

# Fit the model to the training data
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

mse, mae = model.evaluate(X_test, y_test)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
```
